```python
# This command take the inverse of the duplicates... basically grabs all of the values without additional various duplicates.
# Since we're getting rid of rows our indexes will not be entirely sequential, so now we need to add .reset_index()
#drop=True makes it so we won't have an additional index column.
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True).copy()
```

# Asking a question about the data
What are the locations with the fastest rollercoasters? (minimum of 10 rollercoasters at the location)

```python
ax = df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='barh', figsize=(12, 5), title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
